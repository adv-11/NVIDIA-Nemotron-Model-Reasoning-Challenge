```python
import site, os, shutil, torch, safetensors.torch, subprocess, sys, types, importlib
import kagglehub
from peft import LoraConfig, TaskType
from transformers import AutoTokenizer

# ── Paths ─────────────────────────────────────────────────────────────
site.addsitedir("/kaggle/usr/lib/notebooks/ryanholbrook/nvidia-utility-script")
site.addsitedir("/kaggle/usr/lib/notebooks/ryanholbrook/nvidia-utility-script/nvidia_cutlass_dsl/python_packages/")

# ── ptxas ─────────────────────────────────────────────────────────────
os.makedirs("/kaggle/working/bin", exist_ok=True)
for binary, env_var in [
    ("ptxas-blackwell", "TRITON_PTXAS_BLACKWELL_PATH"),
    ("ptxas",           "TRITON_PTXAS_PATH"),
]:
    src = f"/kaggle/input/notebooks/adv1102/nvidia-nemotron-submission/bin/{binary}"
    dst = f"/kaggle/working/bin/{binary}"
    if not os.path.exists(dst):
        shutil.copy2(src, dst)
        os.chmod(dst, 0o755)
    os.environ[env_var] = dst

# ── Mock mamba_ssm ────────────────────────────────────────────────────
fake_mamba = types.ModuleType("mamba_ssm")
fake_mamba.__spec__ = importlib.machinery.ModuleSpec("mamba_ssm", None)
fake_mamba.__version__ = "2.0.4"
fake_mamba.ops = types.ModuleType("mamba_ssm.ops")
fake_mamba.ops.triton = types.ModuleType("mamba_ssm.ops.triton")
fake_mamba.ops.triton.layernorm_gated = types.ModuleType("mamba_ssm.ops.triton.layernorm_gated")
fake_mamba.ops.triton.layernorm_gated.rmsnorm_fn = None
fake_mamba.ops.selective_scan_interface = types.ModuleType("mamba_ssm.ops.selective_scan_interface")
fake_mamba.ops.triton.selective_state_update = types.ModuleType("mamba_ssm.ops.triton.selective_state_update")
fake_mamba.ops.triton.selective_state_update.selective_state_update = None
fake_mamba.ops.triton.ssd_combined = types.ModuleType("mamba_ssm.ops.triton.ssd_combined")
fake_mamba.ops.triton.ssd_combined.mamba_chunk_scan_combined = None
fake_mamba.ops.triton.ssd_combined.mamba_split_conv1d_scan_combined = None
sys.modules.update({
    "mamba_ssm": fake_mamba,
    "mamba_ssm.ops": fake_mamba.ops,
    "mamba_ssm.ops.triton": fake_mamba.ops.triton,
    "mamba_ssm.ops.triton.layernorm_gated": fake_mamba.ops.triton.layernorm_gated,
    "mamba_ssm.ops.selective_scan_interface": fake_mamba.ops.selective_scan_interface,
    "mamba_ssm.ops.triton.selective_state_update": fake_mamba.ops.triton.selective_state_update,
    "mamba_ssm.ops.triton.ssd_combined": fake_mamba.ops.triton.ssd_combined,
})
print("✓ mamba_ssm mock injected")

# ── Just copy adapter files directly — no model loading needed ────────
MODEL_PATH = kagglehub.model_download("metric/nemotron-3-nano-30b-a3b-bf16/transformers/default")
OUTPUT_DIR = "/kaggle/working"
ADAPTER_DATASET = "/kaggle/input/notebooks/adv1102/nvidia-nemotron-submission/nemotron_sft_v5"

# Copy adapter files
for fname in ["adapter_model.safetensors", "adapter_config.json", "chat_template.jinja"]:
    src = f"{ADAPTER_DATASET}/{fname}"
    dst = f"{OUTPUT_DIR}/{fname}"
    if os.path.exists(src):
        shutil.copy2(src, dst)
        print(f"✓ Copied {fname}")

# Save tokenizer
tokenizer = AutoTokenizer.from_pretrained(MODEL_PATH, trust_remote_code=True)
tokenizer.save_pretrained(OUTPUT_DIR)

print(f"✓ Done: {[f for f in os.listdir(OUTPUT_DIR) if not f.startswith('.')]}")
```
